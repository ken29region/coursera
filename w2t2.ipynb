```python
import numpy as np
import math
import scipy.linalg
from matplotlib import pylab as plt
```

```python
def f(x):
    return math.sin(x / 5.) * math.exp(x / 10.) + 5 * math.exp(-x / 2.)

def pl(a, b):
    plt.plot(a, b)
    plt.show()
```

```python
print f(1), f(15)
```

```python
x0 = 1.
y0 = f(x0)#3.0
x1 = 15.
y1 = f(x1)#0.55

a = [[1., x0],
     [1., x1]]
b = [y0, y1]

print scipy.linalg.solve(a, b)

pl(a, b)
```

```python
x1 = 8.
x2 = 15.
y1 = f(x1)
y2 = f(x2)

a = [[1., x0, x0 ** 2], 
     [1., x1, x1 ** 2],
     [1., x2, x2 ** 2]]
b = [y0, y1, y2]

print scipy.linalg.solve(a, b)

pl(a, b)
```

```python
x0 = 1.
x1 = 4.
x2 = 10.
x3 = 15.

y0 = f(x0)
y1 = f(x1)
y2 = f(x2)
y3 = f(x3)

a = [[1., x0, x0 **2, x0 ** 3], 
     [1., x1, x1 **2, x1 ** 3], 
     [1., x2, x2 **2, x2 ** 3], 
     [1., x3, x3 **2, x3 ** 3]]
b = [y0, y1, y2, y3]

c = scipy.linalg.solve(a, b)

print c

pl(a, b)
```

```python
s = ''
for element in c:
    s += '%s%s' % (element, ' ')
print s
```
